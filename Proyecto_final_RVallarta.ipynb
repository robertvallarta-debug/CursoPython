### Biological Aspects of Small Pelagic Fish Associated with Environmental Factors in Northwestern Mexico
## Introduction

Small pelagic fish species—such as sardines, anchovies, and mackerels—constitute a functional group of marine organisms characterized by short life cycles, the formation of large schools, and multiple spawning events throughout the year. These species feed on both phytoplankton and zooplankton and serve as prey for a wide range of intermediate and top-level predators, including numerous species of fish, mollusks, seabirds, and marine mammals. As such, they play a pivotal role in trophic interactions within the epipelagic marine ecosystem.
Their ecological significance lies in their ability to regulate energy flows under the “wasp-waist” trophic control model, which is typical of upwelling systems [1](#ref1). Due to their capacity to reach large biomasses and exhibit broad geographic distribution [2](#ref2), small pelagic fish represent one of the most important fisheries globally, accounting for approximately 20% of total global marine catch [3](#ref3).
In terms of national fisheries, anchovy (Engraulis ringens) comprised over 86% of Peru’s total landed catch volume in 2018. In Chile, of the 2.12 million tons harvested, nearly 60% corresponded to the same species. Overall, the exploitation of various small pelagic fish species contributed to nearly 20% of the global marine fish catch (FAO, 2020). These fisheries are marked by low bycatch rates—non-target species—resulting in relatively high efficiency [4](#ref4), largely due to the schooling behavior that facilitates capture using purse seine nets [5](#ref5).
This research project generated insights into the distributional patterns of small pelagic fish in northwestern Mexico, in relation to environmental factors. These species are of strategic importance for national food security. Expanding our understanding of current climate scenarios and their impacts on marine resources, through the application of innovative technological tools, will enable the implementation of short-term management strategies aimed at sustainable exploitation.


## Hypothesis

Variations in environmental parameters driven by spatiotemporal factors exert significant influence on the biological characteristics of small pelagic fish. These organisms exhibit broad tolerance ranges for certain environmental variables, such as temperature. However, they display narrow tolerance thresholds for others, including salinity and dissolved oxygen, which act as limiting factors for their spatial distribution.
## Objectives.

This study aims to analyze the spatial distribution and abundance patterns of small pelagic fish in northwestern Mexico—specifically in the Gulf of California and the western coast of the Baja California Peninsula—using acoustic survey data. The objective is to determine whether spatial variability in these patterns corresponds to changes in environmental factors during the period from 2019 to 2022.

### Specific objectives.

1. To analyze the size composition of small pelagic fish species.
2. To examine the spatial distribution and abundance of small pelagic fish in northwestern Mexico in order to identify specific patterns related to environmental factors.
3. To assess the environmental parameter ranges associated with the presence of small pelagic fish

### Source of information
The biological and oceanographic information corresponds to research cruises, most of which were conducted to assess the biomass of small pelagic fish in northwestern Mexico, led by the Mexican Institute for Research in Sustainable Fisheries and Aquaculture (IMIPAS) aboard the vessel Dr. Jorge Carranza Fraser.
### Biological sampling
Midwater trawls were carried out using a specialized NMWT 25/25 net designed by Net-System. The hauls were made perpendicular to the coast, with an average duration and speed of 45 minutes and 3.5 knots, respectively. The net has a conical shape consisting of four equal covers (upper and lower selvedge length: 48.17 m or 158 feet) with a 25/25 mesh wing trawl design (vertical/horizontal opening in meters), with mesh sizes decreasing towards the front bag, starting at 1600 mm and ending at 50 mm, made of high-strength materials. The codend is lined with knotless nylon Raschel fabric with a size of 17 mm.
Sampling consisted of simple random sampling, in which the sample volume was approximately 15 kg, obtained randomly from the total volume of the smaller pelagic species that were caught, while still obtaining the total weight of the catch.
Once the small pelagic species had been identified and separated, biological sampling was carried out to obtain a detailed analysis of the organisms, including parameters such as standard and furcal length, total and gutted weight, sex, sexual maturity, gonadal weight, etc. 

During cruises conducted along both the western coast of Baja California and the Gulf of California, a total of seven species of small pelagic fish were caught, providing the information necessary for this study. ### Capture per unit effort (CPUE)
The biomass of D. gigas was estimated using the swept area method, which calculates the area covered by the net during each fishing haul using the following equation:

$$a= t * v * b$$

where a is the swept area (km²), t is the effective trawl time (h), v is the towing speed (km/h), and b is the horizontal opening of the net (m), measured using acoustic sensors mounted on the net during each haul.
Biomass per unit area was calculated using the following equation:

$$Bi= Ci / ai$$

where B₁ is the biomass of sardine caught per unit area (kg/km²) in haul i, C₁ is the total biomass caught in haul i, and a is the swept area.
Catch per unit effort (CPUE, kg/h) was calculated by dividing the biomass (B₁) by the effective trawl time (h) (Sparre & Venema, 1997).

### Environmental Drivers of CPUE and small pelagic fish

Physicochemical data were obtained from 284 oceanographic stations where a previously calibrated Idronaut 320 Plus CTD probe was deployed. Subsequently, measurements of temperature, salinity, and dissolved oxygen were extracted at the depths corresponding to the fishing hauls. To examine the relationship between environmental and biological factors, a Generalized Additive Model (GAM) was implemented. Catch per unit effort (CPUE) and mean mantle length per haul (MML) were used as response variables, while the explanatory variables included longitude (°), latitude (°), seafloor slope (°), bathymetry (m), distance to the edge of the continental shelf (km), temperature (°C), dissolved oxygen (µmol/kg), and salinity. All variables were derived from sampling sites where sardine or anchovy specimens were captured.

The general form of the additive model is as follows (Hastie & Tibshirani, 1986):
$$g(μ)=β_0+f_1 (X_1 )+f_2 (X_2 )+⋯+f_p (X_p )+ε$$

where μ is the response variable, g is the link function, β₀ is the intercept, Xₙ are the predictor variables, fₙ are the smoothing functions, and ε is the error term [7](#ref7).

# Dissolved oxygen
Dissolved oxygen showed higher concentrations in the COPBC than in the GC. As a general reference, it is established that oxygen at sites where the five species were present in the COPBC was above 4 mg/L, with northern anchovy (E. mordax) and Monterey and Japanese sardines being the species that most commonly occupied (statistical modes) areas of higher oxygenation (>6 mg/L). In the GC, the behavior regarding the presence of species in oxygenation conditions was different. For example, as in the previous case, northern anchovy (E. mordax) and Monterey and Japanese sardines showed wide tolerance in oxygenation conditions, from 0.5 to more than 7 mg/L. The Japanese mackerel (S. japonicus) was the most tolerant to hypoxic conditions, as the main mode was 0.9 mg/L. Both the charrito (T. symmetricus) and the crinuda sardine (O. libertate) were mainly found in oxygenated waters (>5 mg/L) with little tolerance to hypoxic conditions. Unlike O. libertate, the crinuda sardine species O. bulleri showed greater tolerance to hypoxic conditions, as it was found at a higher dissolved oxygen threshold (from 0.2 to ~7 mg/L).

# Sea Surface Temperature
Water temperature exhibited warmer conditions in the Gulf of California (GC) compared to the western coast of the Baja California Peninsula (COPBC). As a general reference, the temperature at sites where five species were recorded in the COPBC remained below 15°C. Among these, northern anchovy (Engraulis mordax) and jack mackerel (Trachurus symmetricus) were most frequently associated with colder waters, typically around 10°C, based on modal distribution patterns.
In the GC, northern anchovy, Monterey sardine (Sardinops sagax), Pacific mackerel (Scomber japonicus), and Japanese sardine (Etrumeus acuminatus) similarly favored cooler conditions, commonly occurring in areas with temperatures near 14°C. In contrast, the two threadfin sardine species (Opisthonema libertate and O. bulleri) were found in the warmest waters within the GC, exceeding 25°C—distinct from the thermal preferences of other species. Notably, jack mackerel in the GC, unlike its distribution in the COPBC, was predominantly associated with warmer waters above 20°C.

# Effects of environmental factors on the abundance of small pelagic fish using GAM

In general, small pelagic populations benefit from temperatures between 20 and 27.5 °C, as well as a gradient toward higher salinity (>35 UPS) and a tendency toward lower abundance at higher oxygen concentrations, probably due to the effect of upwelling that affects the environment.  

```python

# Instalar pyGAM si no lo tienes
# !pip install pygam

import pandas as pd
from pygam import LinearGAM, s
import matplotlib.pyplot as plt

# 🔹 Simular un conjunto de datos (puedes reemplazar esto con tus datos reales)
# Variables ambientales (independientes)
import numpy as np
np.random.seed(42)
n = 200
temperature = np.random.uniform(10, 30, n)
salinity = np.random.uniform(30, 38, n)
oxygen = np.random.uniform(4, 8, n)

# Variable biológica (dependiente)
abundance = 10 + 2*np.sin(temperature/3) + 0.5*salinity - 1.5*oxygen + np.random.normal(0, 1, n)

# Crear DataFrame
df = pd.DataFrame({
    'temperature': temperature,
    'salinity': salinity,
    'oxygen': oxygen,
    'abundance': abundance
})

# 🔍 Ajustar el modelo GAM
X = df[['temperature', 'salinity', 'oxygen']]
y = df['abundance']

# Modelo con splines para cada variable independiente
gam = LinearGAM(s(0) + s(1) + s(2)).fit(X, y)

# 📊 Visualizar los efectos de cada variable
fig, axs = plt.subplots(1, 3, figsize=(15, 4))
titles = ['Temperature', 'Salinity', 'Oxygen']
for i, ax in enumerate(axs):
    XX = gam.generate_X_grid(term=i)
    ax.plot(XX[:, i], gam.partial_dependence(term=i, X=XX))
    ax.plot(XX[:, i], gam.partial_dependence(term=i, X=XX, width=0.95)[1], c='r', ls='--')
    ax.set_title(f'Effect of {titles[i]}')
plt.tight_layout()
plt.show()
```
